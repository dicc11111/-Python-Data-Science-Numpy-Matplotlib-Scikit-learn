## Задание 1

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()

boston.keys()
```

```python
data = boston["data"]

data.shape
```

```python
feature_names = boston["feature_names"]

feature_names
```

```python
target = boston["target"]

target[:10]
```

```python
X = pd.DataFrame(data, columns=feature_names)

X.head()
```

```python
X.info()
```

```python
y = pd.DataFrame(target, columns=["price"])

y.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred.shape
```

```python
check_test = pd.DataFrame({
    "y_test": y_test["price"],
    "y_pred": y_pred.flatten(),
})

check_test.head(10)
```

```python
from sklearn.metrics import r2_score
R2 = r2_score(check_test["y_pred"], check_test["y_test"])
R2
```

## Задание 2

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=1000, max_depth=12, random_state=42)
```

```python
model.fit(X_train, y_train.values[:, 0])
```

```python
y_pred = model.predict(X_test)
y_pred.shape
```

```python
check_test = pd.DataFrame({
 "y_test": y_test["price"],
 "y_pred": y_pred.flatten(),
})
check_test.head(10)
```

```python
R2_2 = r2_score(check_test["y_pred"], check_test["y_test"])
R2_2
```

```python
R2_2>R2
```

Модель RandomForestRegressor в данном случае работает лучше, чем LinearRegression

## Задание 4

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('creditcard.csv')
df
```

```python
a = df['Class'].value_counts(normalize=True)
a
```

```python
df.info
```

```python
pd.options.display.max_columns = 100
```

```python
df.head(10)
```

```python
x = df.drop('Class', axis=1)
x
```

```python
y = df['Class']
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=100, stratify=y)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
parameters = {
    'n_estimators': [10, 15], 
    'max_features': np.arange(3, 5), 
    'max_depth': np.arange(4, 7)
}
clf = GridSearchCV(
    estimator=RandomForestClassifier(random_state=100),
    param_grid=parameters,
    scoring='roc_auc',
    cv=3,
)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
clf = RandomForestClassifier(max_depth=6, max_features=3, n_estimators=15)

clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict_proba(X_test)
```

```python
y_pred_proba = y_pred[:, 1]
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_test, y_pred_proba)
```
